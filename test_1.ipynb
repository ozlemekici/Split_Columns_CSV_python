```python
import pandas as pd
import numpy as np
df = pd.read_csv("data_mixed.csv")
df
```

```python
for i in range(df.shape[0]):
    #print(df.iloc[i])
    data = df.iloc[i][1]
    #print(i,":",data)
    data_arr = data.split(",")
    #print(data_arr)
    for keyval in data_arr:
        #print(keyyval)
        if (keyval.find("=")<0):
            continue
        [key,val] = keyval.split("=")
        #print(key,val)
        df.loc[i,key] = val
    #print("--"*15)   

df
```
